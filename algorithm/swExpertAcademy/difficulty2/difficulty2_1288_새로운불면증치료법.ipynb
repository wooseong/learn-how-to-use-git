```python
t = int(input())
for T in range(t):
    n = int(input())
    num=0
    cnt=set()
    while len(cnt)<10 :
        num+=1
        numberList=list(map(int,str(n*num)))
        print(numberList)
        cnt.update(numberList)
        print(cnt)
        print("***********************")
    print(f"#{T+1} {num*n}")
```
