# 조교의 성적 매기기
* https://swexpertacademy.com/main/code/problem/problemDetail.do?problemLevel=2&contestProbId=AV5PwGK6AcIDFAUq&categoryId=AV5PwGK6AcIDFAUq&categoryType=CODE&problemTitle=&orderBy=FIRST_REG_DATETIME&selectCodeLang=ALL&select-1=2&pageSize=10&pageIndex=1

```python
t = int(input())
for T in range(t):
    n,k=map(int, input().split())
    grade=[]
    for N in range(n):
#        mid,final,task = map(int, input().split())
#        grade.append((mid*0.35)+(final*0.45)+(task*0.2))
        grade.append(int(input()))
    print(grade)
    
    kgrade=grade[k-1] #알고자 하는 학생의 성적을 받는다
    print(kgrade)
    
    grade.sort(reverse=True)# 전체 학생의 성적을 정렬한다
    print(grade)
    
    krank=(grade.index(kgrade)+1)/(n/10) # 상대 등수 = 알고자 하는 학생의 절대등수 / (전체 학생이 10의 몇배인지)
    #상대 등수는 전체 학생이 30명일 때 11등(절대등수)이면 10명 기준으로는 4등(상대등수)
    if krank!=int(krank):
        krank+=1
        krank=int(krank)
    print("*********")
    print(grade.index(kgrade)+1)
    print(krank)
    print("*********")
    
    print(f"#{T+1} ",end="")
    if krank==1:
        print("A+")
    elif krank==2:
        print("A0")
    elif krank==3:
        print("A-")
    elif krank==4:
        print("B+")
    elif krank==5:
        print("B0")
    elif krank==6:
        print("B-")
    elif krank==7:
        print("C+")
    elif krank==8:
        print("C0")
    elif krank==9:
        print("C-")
    else :
        print("D0")
```
